## FOOD VENUES IN JAIPUR

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
#Food venues in Jaipur
#GENERATING THE URL TO GET THE FOOD VENUES IN JAIPUR

neighbourhood_latitude=26.916194
neighbourhood_longitude=75.820349
LIMIT = 200 # limit of number of venues returned by Foursquare API
radius = 15000 # define radius
section='food'
# create URL
url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&section={}&limit={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    neighbourhood_latitude, 
    neighbourhood_longitude, 
    radius,
    section,
    LIMIT)
url # display URL
```

### GETTING THE JSON RESULT OF THE URL

```python
results = requests.get(url).json()
results
```

## FILTERING OUT THE DATA AND EXTRACTING THE NAME , CATEGORY AND LATITUDE AND LONGITUDE OF THE RESAUTRANTS

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues
```

## KANHA IS A POPULAR RESTAURANT IN JAIPUR, LETS EXPLORE 

```python
# Exploring popular venues around Kanha Restaurant
latitude=26.903506
longitude=75.753229
```

```python
url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, radius, LIMIT)
url
```

```python
results = requests.get(url).json()
'There are {} around Kanha restaurant.'.format(len(results['response']['groups'][0]['items']))
```

```python
items = results['response']['groups'][0]['items']
items[0]
```

```python
dataframe = json_normalize(items) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories'] + [col for col in dataframe.columns if col.startswith('venue.location.')] + ['venue.id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

# filter the category for each row
dataframe_filtered['venue.categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean columns
dataframe_filtered.columns = [col.split('.')[-1] for col in dataframe_filtered.columns]

dataframe_filtered.head(10)
```

## MAP OF TOP RESTAURANTS AROUND KANHA

#### THE BLACK DOT IS KANHA ON THE MAP

```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start=15) # generate map centred around Ecco


# add Kanha as a violet circle mark
folium.features.CircleMarker(
    [latitude, longitude],
    radius=10,
    popup='KANHA',
    fill=True,
    color='black',
    fill_color='black',
    fill_opacity=0.6
    ).add_to(venues_map)


# add popular spots to the map as blue circle markers
for lat, lng, label in zip(dataframe_filtered.lat, dataframe_filtered.lng, dataframe_filtered.categories):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        fill=True,
        color='red',
        fill_color='blue',
        fill_opacity=0.6
        ).add_to(venues_map)

# display map
venues_map
```
